```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

```python
df = pd.read_csv('processed.csv', index_col=0)
df.head()
```

```python
newYears = []
for y in df['Built/Renov']:
    try:
        if '/' in y:
            newYear = y[-4:]
        else:
            newYear = y 
        newYears.append(newYear)
    except:
        newYears.append(np.NaN)

newYears
```

```python
df['Built/Renov'] = newYears
df.head()
```

```python
df = df[['Units', 'Submarket', 'Built/Renov', 'Vacancy %', 'Parking Spaces/Unit']]
df.head()
```

```python
df.Units.unique()
```

```python
test = '1,500'
test.replace(',','')
```

```python
for col in df.columns:
    if col != 'Submarket':
        if col == 'Units':
            df[col] = [_.replace(',', '') for _ in df[col]]
        df[col] = df[col].astype(float)
```

```python
agg = df.groupby('Submarket').agg({
    'Units': 'mean',
    'Built/Renov': 'mean',
    'Vacancy %': 'median',
    'Parking Spaces/Unit': 'median'
})
agg
```

```python
for col in agg:
    if col in ['Units', 'Built/Renov']:
        agg[col] = agg[col].astype(int)
    else:
        agg[col] = np.round(agg[col], 2)

agg
```

```python
agg.reset_index(inplace=True)
```

```python
agg.to_csv("clean_costar_data.csv", index=False)
```
